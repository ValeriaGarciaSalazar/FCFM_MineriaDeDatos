## Descripción de los datos
- <b>Tipo de datos:</b> Tabla con datos importantes de cada alumno como materias reprobadas, faltas, etc..
- <b>Descripción de las columnas:</b> Nuestra base de datos se conforma por 15 columnas, las cuales se dividen en dos tipos de datos:

 * <b>Binarios:</b> Sexo, Entorno, Fam, Estatus_padres, Extracurriculares, Universidad
 * <b>Numérico:</b> Edad, Tiempo_estudio, Reprobadas, Famrel, Tiempo_libre, Salir, Dalc, Falc, Faltas
 
- ### Descripción por columna:

 * <b>Sexo:</b> Sexo del estudiante (F "Femenino" o M "Masculino")
 * <b>Edad:</b> Edad del estudiante [15,22]
 * <b>Entorno:</b> Zona en la que vive el estudiante (U "Urbano" o R "Rural")
 * <b>Fam:</b> Número de integrantes en la familia ("3 o -" menor o igual a 3 o "4 o +" más de 3)
 * <b>Estatu_padres:</b> Cómo viven los padres (S "Separados" o J "Juntos")
 * <b>Tiempo_estudio:</b> Horas de estudio a la semana ([1,4] donde 1=1-2 hrs, 2=2-5 hrs, 3=5-10 hrs y 4=10 o más hrs)
 * <b>Reprobadas:</b> Número de clases que ha reprobado [0,3]
 * <b>Extracurriculares:</b> Actividades fuera del horario escolar ("Si" o "No")
 * <b>Universidad:</b> Quiere cursar universidad acabando la preparatoria  ("Si" o "No")
 * <b>Famrel:</b> Calidad de su relación familiar (De 1=Muy mala a 5=Excelente)
 * <b>Tiempo_libre:</b> Tiempo libre que tiene (De 1=Muy poco a 5=Mucho)
 * <b>Salir:</b> Frecuencia en que sale con amigos (De 1=Muy poco a 5=Mucho)
 * <b>Dalc:</b> Consumo de alcohol entre semana (De 1=Muy poco a 5=Mucho)
 * <b>Falc:</b> Consumo de alcohol el fin de semana (De 1=Muy poco a 5=Mucho)
 * <b>Faltas:</b> Número de faltas [0,93]

## Gráficos [Vizualización de Datos]

```python
import pandas as pd 
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
bd= "C:/Users/anavi/Desktop/base_final.csv"
base= pd.read_csv(bd, header = 0)
```

- ### Gráfico de Barras

```python
sns.barplot(x=base['Edad'],y=base['Falc'])
plt.title("Cantidad de alcohol el fin de semana")
```

Con el gráfico anterior, podemos observar la cantidad de alcohol que consumen los alumnos de edad x los fines de semana.

Con el gráfico anterior podemos observar que hay mas mujeres que hombres
